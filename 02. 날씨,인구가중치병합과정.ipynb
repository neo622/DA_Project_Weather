## 수집한 지역 별 인구 가중치 데이터를 판매 수량 데이터에 적용한다.

```python
buy2018_1 = pd.read_csv('C:/Users/admin/Desktop/mini/buy2018_1.csv',engine='python')
buy2018_2 = pd.read_csv('C:/Users/admin/Desktop/프로젝트 데이터/원본데이터/buy2018_2.csv',engine='python')
buy2019_1 = pd.read_csv('C:/Users/admin/Desktop/프로젝트 데이터/원본데이터/buy2019_1.csv',engine='python')
buy2019_2 = pd.read_csv('C:/Users/admin/Desktop/프로젝트 데이터/원본데이터/buy2019_2.csv',engine='python')
```

```python
buy2018_1.columns = ['a','date','sex','age','big_cat','sm_cat','qty']
buy2018_2.columns = ['a','date','sex','age','big_cat','sm_cat','qty']
buy2019_1.columns = ['a','date','sex','age','big_cat','sm_cat','qty']
buy2019_2.columns = ['a','date','sex','age','big_cat','sm_cat','qty']
```

```python
total = pd.concat([buy2018_1,buy2018_2,buy2019_1,buy2019_2])  ## 2년치 데이터 합친 것
```

```python
total = total.drop('a',axis = 1)  ## 쓸모없는 열 제거
```

```python
total_a = total.groupby(by=['date','big_cat','sm_cat'], as_index=False).sum()  ## 성별, 나이 신경안쓰고 qty만 집중
```

```python
total_a
```

```python
total_b = total_a.drop(['big_cat','age'],axis=1)  ## 쓸모없는 열 제거
```

```python
total_qty = np.array(total_b['qty']).reshape(267707,1)
total_qty
```

```python
area_weight = pd.read_excel('C:/Users/admin/Desktop/프로젝트 데이터/area_weight.xlsx')
columns = area_weight.area.unique()
weight=area_weight['weight']
weight_1 = weight.tolist()
```

```python
weight_1 = np.array(weight_1)
```

```python
weight_1= weight_1.reshape(1,16)
```

```python
weight_1                                 ## 지역별 인구가중치와 qty 곱해주기 위한 과정
```

```python
#가중치곱한 데이터프레임 a
a = np.dot(total_qty,weight_1)
b= pd.DataFrame(a)
b.columns = columns                     ## b가 원하는 데이터프레임
```

```python
k = pd.merge(total_b,b,left_on=total_b.index,right_on = total_b.index)
```

```python
k   ## 합치기 성공
```

```python
k.drop('key_0',inplace=True,axis=1)     # 쓸모없는 열 제거
```

```python
k.to_csv('날씨에따른수요예측.csv', index=False)   #저장
```

```python
import numpy as np
import pandas as pd
se_weather = pd.read_csv('C:/Users/admin/Desktop/프로젝트 데이터/날씨+미세먼지/서울데이터.csv',engine='python')

week =[]
value = 0
for i in range(730):
    
    week.append(value)
    if value > 5:
        value = 0
    else:
        value+=1
## week data추가
se_weather['week'] = week 

from sklearn.preprocessing import OneHotEncoder
enc= OneHotEncoder()
enc.fit(se_weather.week.values.reshape(-1,1))

week_ohe = enc.transform(se_weather.week.values.reshape(-1,1)).toarray()
week_ohe = pd.DataFrame(week_ohe)
week_ohe.columns = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
X = pd.concat([se_weather,week_ohe], axis = 1)
se_weather_week = X.drop('week',axis=1)
```
