## **물품에 따른 mse, r2, coefficient**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
```

```python
reg_test = []
for sm_cat in total_incheon.sm_cat.unique():
    rf_model = RandomForestRegressor()
    df = total_incheon.loc[total_incheon.sm_cat == sm_cat,:]         #### 물품으로 group화하여 각 물품마다 mse,r2,coefficient 구하기
    X = df.iloc[:,3:]
    y = df.iloc[:,2]
    rf_model.fit(X,y)
    mean_squared_err = mse(y,rf_model.predict(X))
    r2 = rf_model.score(X,y)
    
    lm = smf.ols(formula = '인천광역시 ~ tem + rain + hum + snow + dust + Mon + Tue + Wed + Thu + Fri + Sat + Sun', data = df)
    result = lm.fit()
    a = result.conf_int().mean(axis=1)          ## 각 coefficient
    
    tmp = (sm_cat, format(mean_squared_err,'6f'), format(r2,'6f'), format(a.values[0],'6f'), format(a.values[1],'6f'), 
          format(a.values[2],'6f'), format(a.values[3],'6f'), format(a.values[4],'6f'), format(a.values[5],'6f'), 
          format(a.values[6],'6f'), format(a.values[7],'6f'), format(a.values[8],'6f'), format(a.values[9],'6f'), 
          format(a.values[10],'6f'), format(a.values[11],'6f'), format(a.values[12],'6f'))
    reg_test.append(tmp)
```

```python
mrc_inc = pd.DataFrame(reg_test)       ## mrc (mse, r2, coef 줄임)
```

```python
mrc_inc = pd.DataFrame(reg_test)
mrc_inc.columns=['sm_cat','MSE','R2_score','Intercept_coef','tem_coef','rain_coef','hum_coef','snow_coef','dust_coef',
                 'Mon_coef','Tue_coef','Wed_coef','Thu_coef','Fri_coef','Sat_coef','Sun_coef']
## column 이름 변경
```
