```python
@app.callback(
    Output('pie-graph','figure'),
    [Input('tem-dropdown','value'),
    Input('dust-dropdown','value'),
    Input('hum-dropdown','value'),
    Input('age-dropdown','value'),
    Input('sex-dropdown','value'),
    Input('big-radioitem','value')]
)
def pie_graph_total(tem,dust,hum,age,sex,big):
    date = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) \
                          & (df_weather.hum_a == hum),'date']
    area = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) \
                          & (df_weather.hum_a == hum),'area'].unique()
    area_add = np.append(area, np.array(['middle','age','sex','big']))
    index = df_data.loc[df_data.date.isin(date),'index']
    data = df_data[area_add].loc[index]
    data['qty'] = data.sum(axis=1)
    data1 = data.loc[(data.age == age) & (data.sex == sex) & (data.big == big), \
                     ['middle','qty']].groupby('middle').sum().reset_index()
    fig = px.pie(data1, values=data1.qty, names=data1.middle ,
                color_discrete_sequence=px.colors.sequential.RdBu,
                template = 'plotly_dark')    
    return fig


@app.callback(
    Output('line-graph','figure'),
    [Input('tem-dropdown','value'),
    Input('dust-dropdown','value'),
    Input('hum-dropdown','value'),
    Input('age-dropdown','value'),
    Input('big-radioitem','value')]
)
def bar_graph_age(tem,dust,hum,age,big):
    date = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) & \
                          (df_weather.hum_a == hum),'date']
    area = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) & \
                          (df_weather.hum_a == hum),'area'].unique()
    area_add = np.append(area, np.array(['middle','age','big']))
    index = df_data.loc[df_data.date.isin(date),'index']
    data = df_data[area_add].loc[index]
    data['qty'] = data.sum(axis=1)
    data1 = data.loc[(data.age == age) & (data.big == big),['middle','qty']].groupby('middle').sum().reset_index()
    fig = px.bar(data1, x=data1.middle, y=data1.qty, template = 'plotly_dark', text = data1.qty)
    colors = ['#b22222','#191970','#ffffff','#ffe4c4','#b0e0e6','#800000','#4169e1','#ffa07a'] * 4
    fig.update_traces(marker_color = colors)
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(uniformtext_minsize=6, uniformtext_mode='show')
    fig.update_layout(title="날씨에 따른 연령별 소비트렌드", 
                  font_family="San Serif",
                  titlefont={'size': 16},
                  legend=dict(
                  orientation="v", y=1, yanchor="top", x=1.25, xanchor="left")
                  ).update_xaxes(categoryorder='total descending')
    fig.update_xaxes(title_text = '상품')
    fig.update_yaxes(title_text = '구매건수')
    return fig

@app.callback(
    Output('line-graph2','figure'),
    [Input('tem-dropdown','value'),
    Input('dust-dropdown','value'),
    Input('hum-dropdown','value'),
    Input('sex-dropdown','value'),
    Input('big-radioitem','value')]
)
def bar_graph_sex(tem,dust,hum,sex,big):
    date = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) & \
                          (df_weather.hum_a == hum),'date']
    area = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) & \
                          (df_weather.hum_a == hum),'area'].unique()
    area_add = np.append(area, np.array(['middle','sex','big']))
    index = df_data.loc[df_data.date.isin(date),'index']
    data = df_data[area_add].loc[index]
    data['qty'] = data.sum(axis=1)
    data1 = data.loc[(data.sex == sex) & (data.big == big),['middle','qty']].groupby('middle').sum().reset_index()
    fig = px.bar(data1, x=data1.middle, y=data1.qty, template = 'plotly_dark', text = data1.qty)
    colors = ['#b22222','#191970','#ffffff','#ffe4c4','#b0e0e6','#800000','#4169e1','#ffa07a'] * 4
    fig.update_traces(marker_color = colors)
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(uniformtext_minsize=6, uniformtext_mode='show')
    fig.update_layout(title="날씨에 따른 성별 소비트렌드", 
                  font_family="San Serif",
                  titlefont={'size': 16},
                  legend=dict(
                  orientation="v", y=1, yanchor="top", x=1.25, xanchor="left")
                  ).update_xaxes(categoryorder='total descending')
    fig.update_xaxes(title_text = '상품')
    fig.update_yaxes(title_text = '구매건수')
    return fig

@app.callback(
    Output('pie-graph2','figure'),
    [Input('tem-dropdown','value'),
    Input('dust-dropdown','value'),
    Input('hum-dropdown','value'),
    Input('middle-dropdown','value')]
)
def pie_graph_product(tem,dust,hum,middle):
    area = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) & (df_weather.hum_a == hum),'area'].unique()
    date = df_weather.loc[(df_weather.tem_range == tem) & (df_weather.dust_a == dust) & (df_weather.hum_a == hum),'date']
    area_add = np.append(area, np.array(['middle']))
    index = df_data.loc[df_data.date.isin(date),'index']
    data = df_data[area_add].loc[index]
    data['qty'] = data.sum(axis=1)
    data1 = data.loc[:,['middle','qty']].groupby('middle').sum().reset_index()
    a = df_data.loc[:,['qty','middle']].groupby('middle').sum()
    b = np.array(data1.loc[:,'middle'])
    aa = a.loc[b,['qty']].reset_index()
    data1['qty_r'] = aa.qty - data1.qty
    data2 = data1.loc[data1.middle==middle,['qty','qty_r']]
    data3 = data2.T
    data3['판매여부']=['판매량',' ']
    data3.columns = ['판매량','여부']
    fig = px.pie(data3, values=data3.판매량, names=data3.여부,
                 color_discrete_map={'비판매':'#0000cd','판매':'#00bfff'},
                 template = 'plotly_dark', hole = .7)
    #fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(annotations=[dict(text=str(np.round(data3.판매량[0]/(data3.판매량[1]+data3.판매량[0])*100,1)) + '%', x=0.5, y=0.5, \
                                        font_size=20, showarrow=False)])
    fig.update_layout(showlegend=False)
    return fig
```
