```python
@app.callback(
    Output('line-graph','figure'),
    [Input('area-dropdown','value'),
    Input('cat-dropdown','value'),
    Input('variable-dropdown','value')
    ]
)
def graph(area,smcat,factor):
    data = master_area_weather1.loc[(master_area_weather1.area == area) & (master_area_weather1.sm_cat == smcat)][[factor,'qty_area']]
    fig_line = px.bar(data, x=data[factor], y='qty_area')
    
    return fig_line

@app.callback(
    Output('predict-output','children'),
    [Input('area-dropdown','value'),
    Input('cat-dropdown','value'), 
    Input('tem-input','value'),
    Input('rain-input','value'),
    Input('hum-input','value'),
    Input('snow-input','value'),
    Input('dust-input','value'),
    Input('mon-input','value'),
    Input('tue-input','value'),
    Input('wed-input','value'),
    Input('thu-input','value'),
    Input('fri-input','value'),
    Input('sat-input','value'),
    Input('sun-input','value')
    ]
)
def number_render(area,smcat,tem,rain,hum,snow,dust,mon,tue,wed,thu,fri,sat,sun):
    rf_model = RandomForestRegressor()
    df = master_area_weather1.loc[(master_area_weather1.area == area) & (master_area_weather1.sm_cat == smcat)]
    X = df.iloc[:,4:16]
    y = df.iloc[:,3]
    rf_model.fit(X,y)
    qty_predict = [[tem,rain,hum,snow,dust,mon,tue,wed,thu,fri,sat,sun]]  ### input 
    
    

    return print(rf_model.predict(qty_predict),"개")
```

```python
reg_test = []
for sm_cat in master_area_weather1.sm_cat.unique():
    rf_model = RandomForestRegressor()
    df = master_area_weather1.loc[master_area_weather1.sm_cat == sm_cat,:]         #### 물품으로 group화하여 각 물품마다 mse,r2,coefficient 구하기
    X = df.iloc[:,4:16]
    y = df.iloc[:,3]
    rf_model.fit(X,y)
    mean_squared_err = mse(y,rf_model.predict(X))
    r2 = rf_model.score(X,y)
    
    lm = smf.ols(formula = 'qty_area ~ tem + rain + hum + snow + dust + Mon + Tue + Wed + Thu + Fri + Sat + Sun', data = df)
    result = lm.fit()
    a = result.conf_int().mean(axis=1)          ## 각 coefficient
    
    tmp = (sm_cat, format(mean_squared_err,'6f'), format(r2,'6f'), format(a.values[0],'6f'), format(a.values[1],'6f'), 
          format(a.values[2],'6f'), format(a.values[3],'6f'), format(a.values[4],'6f'), format(a.values[5],'6f'), 
          format(a.values[6],'6f'), format(a.values[7],'6f'), format(a.values[8],'6f'), format(a.values[9],'6f'), 
          format(a.values[10],'6f'), format(a.values[11],'6f'), format(a.values[12],'6f'))
    reg_test.append(tmp)
```
